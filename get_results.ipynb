```python
from time import sleep
import pandas as pd
import json

# sleep(36000)
```

```python
with open('temp_res.txt', 'r', encoding='utf-8') as f:
    data = []
    for line in f:
        try:
            data.append(json.loads(line))
        except:
            pass
```

```python
data = list(filter(lambda x: type(x['data']) == list, data))
```

```python
df = pd.DataFrame(columns=['id', 'subject_pre', 'city_pre', 'descr', 'subject', 'subject_status', 'region', 'region_status', 'city', 'city_status', 'street', 'ststatus', 'full'])
```

```python
c, l = 1, len(data)
for rec in data:
    for d in rec['data']:
        df = df.append(pd.Series({
            'id': rec.get('id'),
            'subject_pre': rec.get('subject'), 
            'city_pre': rec.get('city'), 
            'descr': rec.get('descr'), 
            'subject': d.get('subject'), 
            'subject_status': d.get('sstatus'), 
            'region': d.get('mo'), 
            'region_status': d.get('mostatus'), 
            'city': d.get('city'), 
            'city_status': d.get('cstatus'), 
            'street': d.get('street'),
            'ststatus': d.get('ststatus'),
            'full': d.get('full')
        }), ignore_index=True)
    c += 1
    if not c % 100:
        print(f'{c}/{l}')
```

```python
df.to_excel('analyze.xlsx')
```
